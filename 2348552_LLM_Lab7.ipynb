```python
import pandas as pd
from transformers import BertTokenizer
```

```python
data = {
    'text': [
        "Patient reports severe headaches and dizziness.",
        "The prescription includes ibuprofen and paracetamol for pain relief.",
        "The patient needs to undergo an MRI scan next week.",
        "Blood tests indicate elevated levels of cholesterol."
    ]
}
```

```python
df = pd.DataFrame(data)
```

```python
# Initialize the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
# Tokenize the text data
df['tokens'] = df['text'].apply(lambda x: tokenizer.tokenize(x))
print(df[['text', 'tokens']])
```

```python
# Encode the text data
df['input_ids'] = df['text'].apply(lambda x: tokenizer.encode(x, add_special_tokens=True))

# Display the tokens and their corresponding input IDs
print(df[['text', 'tokens', 'input_ids']])
```
